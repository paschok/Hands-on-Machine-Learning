### from <b>Housing</b> where we built corr_matrix I made up my mind that <i>total_bedrooms is not important</i>

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.3, random_state=42)
```

```python
def drop_attr(df, attribue):
    return df.drop(attribue, axis=1)
```

```python
train_set_no_bed = drop_attr(train_set, 'total_bedrooms')
train_set_no_bed
```

```python
X_train = drop_attr(train_set_no_bed, 'median_house_value')
X_train
```

```python
y_train = train_set_no_bed["median_house_value"].copy()
y_train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer 
```

```python
list(X_train)
```

```python
num_attribs = list(drop_attr(X_train, "ocean_proximity"))
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([('std_scaler', StandardScaler(), num_attribs), 
                                   ("cat", OneHotEncoder(), cat_attribs),])
```

```python
housing_prepared = full_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression() 
lin_reg.fit(housing_prepared, y_train)
```

```python
test_set_no_bed = drop_attr(test_set, 'total_bedrooms')
X_test = drop_attr(test_set_no_bed, 'median_house_value')
X_test_prepared = full_pipeline.fit_transform(X_test)
```

```python
y_pred = lin_reg.predict(X_test_prepared)
```

```python
y_test = test_set_no_bed["median_house_value"].copy()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
```
